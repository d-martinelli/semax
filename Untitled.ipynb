```python
from Bio import PDB
import numpy as np

parser = PDB.PDBParser()
io = PDB.PDBIO()
struct = parser.get_structure('semax','pepfoldsemax.pdb')

translation = array((66.389, 80.366, 48.833), 'f')

rotation_matrix = PDB.rotmat(PDB.Vector([0, 0, 0]), PDB.Vector([0, 0, 0]))

# for atom in struct.get_atoms():
#     atom_C1 = atom.coord.copy()
#     break

for model in struct:
    for chain in model:
        for residue in chain:
            for atom in residue:
                atom.transform(rotation_matrix, translation)

io.set_structure(struct)
io.save('semax_coord.pdb')
```

```python

from Bio.PDB import PDBParser, PDBIO

io = PDBIO()
pdb = PDBParser().get_structure("no_mcr4complex", "no_mcr4complex.pdb")

for chain in pdb.get_chains():
    if chain.get_id() == 'L':
        io.set_structure(chain)
        io.save(pdb.get_id() + "_" + chain.get_id() + ".pdb")
```
